```python
N_trials = 10

t = (np.random.random(N_trials)*100)

ML = t
MAP = 2*t

plt.hist(ML,bins = 'auto')
plt.title("ML est")
plt.show()
plt.hist(MAP,bins = 'auto')
plt.title("MAP est")
plt.show()
```

```python
def Lbus(tau,t):
    i = 0
    out  = 1 
    while (i < np.size(t)):
        if (t[i] > tau):
            temp = 0
        else :
            temp = t[i] / tau**2
        out*=temp
        i+=1
    return(out)

print(Lbus(50,t))


N = 150
tau = (np.arange(N)+1)

L = np.zeros(N)

i = 0
while (i < N):
    L[i] = Lbus(tau[i],t)
    i+=1


plt.plot(tau,L)
```
